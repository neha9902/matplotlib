```python
plt.plot([1,2,3],[5,7,1])
plt.title("Data")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x1 = [1,2,3]
y1 = [1,2,3]
x2 = [1,2,3]
y2 = [3,2,1]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
x1 = [1,2,3]
y1 = [1,2,3]
x2 = [1,2,3]
y2 = [3,2,1]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
from matplotlib import style
plt.style.use('grayscale')
x1 = [1,2,3]
y1 = [1,2,3]
x2 = [1,2,3]
y2 = [3,2,1]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
x1 = [1,2,3]
y1 = [1,2,3]
x2 = [1,2,3]
y2 = [3,2,1]
plt.plot(x1,y1, label ='first')
plt.plot(x2,y2, label ='second')
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x1 = [1,2,3]
y1 = [1,2,3]
x2 = [1,2,3]
y2 = [3,2,1]
plt.plot(x1,y1,label ='first',linewidth=3)
plt.plot(x2,y2,label ='second')
plt.legend()
plt.show()
```
